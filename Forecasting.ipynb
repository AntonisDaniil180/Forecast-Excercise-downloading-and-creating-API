```python
import requests
import pandas as pd
import numpy as np
```

```python
# taking London APIs with concat and appending different data in a single for loop

# adding rangeminL and rangemaxL to choose the range you want

# adding month year also to be able to choose month and year for a location you want
month=9

year=2021

rangeminL=13

rangemaxL=21

London = []

for i in range(rangeminL+1,rangemaxL):
    if i<10:
        url='https://www.metaweather.com/api/location/44418/'+str(year)+'/'+str(month)+'/0'+str(i)+'/'
    else:
        url='https://www.metaweather.com/api/location/44418/'+str(year)+'/'+str(month)+'/'+str(i)+'/'
    res = pd.DataFrame(requests.get(url).json())    
    London.append(res)

# reseting index
London = pd.concat(London).reset_index()
London.drop('index',axis=1,inplace=True)

# adding an extra column to identify the location
London['location']='London UK'
 
London
```

```python
# taking San Fransisco APIs with concat and appending different data in a single for loop
rangeminSF=13
rangemaxSF=21
SanFr = []

for i in range(rangeminSF+1,rangemaxSF):
    if i<10:
        url='https://www.metaweather.com/api/location/2487956/'+str(year)+'/'+str(month)+'/0'+str(i)+'/'
    else:
        url='https://www.metaweather.com/api/location/2487956/'+str(year)+'/'+str(month)+'/'+str(i)+'/'
        
    res = pd.DataFrame(requests.get(url).json())    
    SanFr.append(res)
# reseting index
SanFr = pd.concat(SanFr).reset_index()
SanFr.drop('index',axis=1,inplace=True)
SanFr['location']='San Fransisco USA'
SanFr
```

```python
# taking San Antonio APIs with concat and appending different data in a single for loop
rangeminSA=13
rangemaxSA=21
SanAn = []
for i in range(rangeminSA,rangemaxSA):
    if i<10:
        url='https://www.metaweather.com/api/location/2487796/'+str(year)+'/'+str(month)+'/0'+str(i)+'/'
    else:
        url='https://www.metaweather.com/api/location/2487796/'+str(year)+'/'+str(month)+'/'+str(i)+'/'
    res = pd.DataFrame(requests.get(url).json())    
    SanAn.append(res)
# reseting index
SanAn = pd.concat(SanAn).reset_index()
SanAn.drop('index',axis=1,inplace=True)
SanAn['location']='San Antonio USA'
SanAn
```

```python
# concat all data in a single file

all_data=pd.concat([London,SanFr,SanAn]).reset_index()
all_data.drop('index',axis=1,inplace=True)
```

```python
# some data cleaning/formating because of decimals

# temperature: I would round to decimal = 3 so I wont miss any valuable information

# wind_direction: should be rounded to decimal = 3 because it is calculated to degrees

# for visibility I would keep the null values because they represent that we have no information

for i in ['min_temp','max_temp','the_temp','wind_direction','visibility']:
    all_data.round(decimals=3)
    
# wind_speed should be rounded to decimal = 2 because it is calculated to mph

for i in ['wind_speed','air_pressure']:
    all_data.round(decimals=2)
     
# humidity it is a percentage as it is stated in https://www.metaweather.com/api/ so i round to 0

# predictability would keep the Null values 

for i in ['humidity','predictability']:
    all_data.round(decimals=0)

all_data    

```

```python
# changing created column in order to be readable as datetime
all_data['created']= all_data['created'].apply(lambda x: x.split('T')[0]+ " " + x.split('T')[1].split('.')[0])

# getting our data to csv and then add it to our relational database (check MySQL file)
all_data.to_csv('all_data.csv')
all_data
```

# Creating an API for latest forecast 'created' for each location and each day 
# (groupby(['location','applicable_date'])

```python
#when you run the cell this should be the address where the endpoints are being shown http://127.0.0.1:7433/
app= Flask(__name__)
@app.route('/',methods=['GET'])

def request_page():
    
    maximum=df.groupby(['location','applicable_date'])['created'].max().reset_index()
    
#   convert our data to dictionary so it could be transformed to json file
    data_set=maximum.to_dict()
    
    json_dump= json.dumps(data_set)
    
    return json_dump
if __name__=='__main__':
    app.run(port=7433)
```

# List the average temp of the last 3 forecasts for each location for every day

```python
#the whole API address is http://127.0.0.1:5555/random

app= Flask(__name__)

new=df
@app.route('/random',methods=['GET'])
def page():
    
    # taking only the date from forecast in order to group by
    new['created']=new['created'].apply(pd.to_datetime)
    
    new['date_of_forecast']=new['created'].apply(lambda x: x.date())
    
    # using dense_rank(in case we have multiple top 3) to get the top 3 for each day for every location
    
    new['ranked']=new.groupby(['location','date_of_forecast'])['created'].rank(method='dense',ascending=False)
    
    # using mean function to take the average of the last 3 the_temp
    
    requested=new[new['ranked']<=3].groupby(['location','date_of_forecast'])['the_temp'].mean().reset_index()
    
    # convert date_of_forecast into string since dictionary cant read datetime types
    
    requested['date_of_forecast']=requested.date_of_forecast.astype(str)
    
    data_new=requested.to_dict()
    
    json_dump= json.dumps(data_new)
    return json_dump
if __name__=='__main__':
    app.run(port=5555)
```

# Get the top n locations based on each available metric where n is a parameter given to the API call.

```python
# importing our data from the created csv file because of an error that some columns 
# are datetime even though we changed them
df=pd.read_csv(r'C:\Users\adoni\Desktop\margera\all_data.csv')

# dropping unnamed column
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
app= Flask(__name__)

x=df

# making a list of metrics for best practice

metrics = list(df.columns)[6:15] 

         
@app.route('/new/<n>') 
# So the location would be like  http://127.0.0.1:4424/new/n
# For example the url address for the first 10 values would look like this: http://127.0.0.1:4424/new/10

def requested_page(n):        
    
    datafile= []
    
#   sorting by each available metric by using sort_values(by=cl_names[i],ascending=False)

#   ascending=False to get the top values

#   python read id as a string by int(n) we turn it into an integer

#  taking only the metrics from 6-15

    for m in metrics:
        
            x=df.sort_values(by=m,ascending=False).head(int(n))    
            
            datafile.append(x)
            
    datafile= pd.concat(datafile)   
    
    data_set=datafile.to_dict() 
    
    json_dump= json.dumps(data_set)   
    
    return json_dump

if __name__=='__main__':
    app.run(port=4433)
```
